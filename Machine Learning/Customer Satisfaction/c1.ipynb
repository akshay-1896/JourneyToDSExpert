## One Hot Encoding through Pandas

```python
cat_col = df.select_dtypes(include="O")
num_col = df.select_dtypes(exclude="O")
```

```python
cat_col.head()
```

```python
pd_encoded_data = pd.get_dummies(cat_col, drop_first=True).astype(int)
```

```python
complete_df = pd.concat([pd_encoded_data, num_col], axis=1)
complete_df.head()
```

```python
complete_df.dtypes
```

```python
# Data Splitting

x = complete_df.drop("satisfaction_satisfied", axis=1)
x.head()
x['Flight Distance'].max()
```

```python
y = complete_df[["satisfaction_satisfied"]]
y.head()
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.columns
```

```python
# Model Selection
from sklearn.linear_model import LogisticRegression
```

```python
lgr = LogisticRegression(max_iter=1000)
```

```python
# Model Training

lgr.fit(x_train, y_train)
```

```python
# Model Testing

y_pred = lgr.predict(x_test)
score = lgr.score(x_test, y_test)
score
```

```python
# y_test ==> actual values
# y_pred ==> predicted values
matrix = confusion_matrix(y_test, y_pred)
matrix
```
